````
def CreateDownloadArea(file='') : 
  
  #path = Path('data/planes')
  dest = path/file
  dest.mkdir(parents=True, exist_ok=True)
  
  import os
  os.system('cp ' + file +'.txt' + ' ' + 'data/planes/'+file)
  
  classes.append(file)
  
  download_images('data/planes/' + file + '/' + file +'.txt' , 'data/planes/'+file, max_pics=200)
````

# Data Viewing and making training set

````
np.random.seed(42)
data = ImageDataBunch.from_folder(path, train=".", valid_pct=0.2,
        ds_tfms=get_transforms(), size=224, num_workers=4).normalize(imagenet_stats)
````

````
data
````

````
data.classes
````

````
data.show_batch(rows=7, figsize=(12,16))
````

````
data.classes, data.c, len(data.train_ds), len(data.valid_ds)
````

# Training the image data with some specific model 

````
learn = cnn_learner(data, models.resnet34, metrics=error_rate)
````

````
#learn
````

# Let's take an image and see what the pre-trained model is saying 

````
test_image = learn.data.valid_ds[15]
test_image
````

````
test_image[0]
````

````
prediction = learn.predict(test_image[0])
````

````
prediction
````

````
data.classes
````

````
import matplotlib.pyplot as plt
plt.bar( x = np.arange(4), height = prediction[2].data.numpy() )
plt.xticks(np.arange(0, 4, step=1))
plt.ylim(0,1)
plt.show()
````

````
learn.fit_one_cycle(2)
````

````
prediction1 = learn.predict(test_image[0])
````

````
import matplotlib.pyplot as plt
plt.bar( x = np.arange(4), height = prediction1[2].data.numpy() )
plt.xticks(np.arange(0, 4, step=1))
plt.ylim(0,1)
plt.show()
````

````
learn.save('stage-1-resnet34') # -- saving the trained model --- #
````
